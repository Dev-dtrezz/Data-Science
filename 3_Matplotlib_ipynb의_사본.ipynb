```python
plt.rc('font', family='NanumBarunGothic') # 폰트 적용시키는 코드
```

```python
plt.figure(figsize=(6, 8)) # inch
plt.plot([1,2,3],[1,2,3])
plt.plot([1,2,3],[2,4,6])
plt.title('제목', fontsize=30) # pixel
plt.xlabel('X축', fontsize=20)
plt.ylabel('Y축', fontsize=20, rotation = 30)
plt.show()
```

```python
plt.figure(figsize=(15, 20)) # inch / 사이즈지정
plt.title('마커설정', fontsize=30) # pixel
plt.plot(np.arange(10), np.arange(10), color='deepskyblue', marker='o', linestyle='-.', ms=10)
plt.plot(np.arange(10), np.arange(10)*2, color='deeppink', marker='v', linestyle='--')
plt.plot(np.arange(10), np.arange(10)*3, color='gold', marker='*', linestyle='')

# 범례: 기본위치는 왼쪽 상단
plt.legend(['10', '10*2','10*3'], fontsize=15, loc='lower right', ncol=3)

plt.xlim(0,12) # x축의 리미트를 지정해줌(범위)
plt.ylim(0,30)

plt.xticks(rotation=30) # x축 단위값들 회전
plt.yticks(rotation=45)

plt.grid() #그리드 추가

plt.show()
```

```python
x = ['파이썬', '데이터분석', '머신러닝', '딥러닝', '자연어처리', '컴퓨터비전']
y = [95, 80, 75, 50, 40, 25]
plt.figure(figsize=(8,5))
plt.title('AI성적표', fontsize=25)
plt.ylabel('수강생 점수')
plt.bar(x, y, alpha=0.4, color = 'deeppink') # alpha는 투명도
plt.show()
```
