```python
# 각 컬럼당 null이 몇 개 있는지 확인
retail.isnull().sum()
```

```python
# 비회원 / 탈퇴 / 휴면회원 제거
retail = retail[pd.notnull(retail['CustomerID'])]
retail
```

```python
len(retail)
```

```python
retail.describe() # 전체 통계를 보면서 잘못된 데이터 구문(ex. 음수, 0값)
```

```python
# 구입 수량이 0이하인 데이터를 확인 및 제거
retail[retail['Quantity'] <= 0]
```

```python
# 구입 수량이 1이상인 데이터만 저장
retail = retail[retail['Quantity'] >= 1]
len(retail)
```

```python
# 구입 가격이 0 이하인 데이터를 확인
retail[retail['UnitPrice'] <= 0]
```

```python
# 구입 가격이 0초과인 데이터만 저장
retail = retail[retail['UnitPrice']>0]
len(retail)
```

```python
# 고객의 총 지출비용(CeckoutPrice) 파생변수 만들기
# 총 지출비욜 = 가격(UnitPrice) * 수량(Quantity)
retail['CheckoutPrice'] = retail['UnitPrice'] * retail['Quantity']
retail
```

```python
retail.info()
```

```python
retail.head()
```

```python
# 주문날짜 데이터 : 날짜타입으로 바꾸기
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
retail.info()
```

```python
# 전체 매출
total_revenue = retail['CheckoutPrice'].sum()
total_revenue
```

```python
# 각 나라별 구매 횟수
retail['Country'].value_counts()

# pd.options.display.max_rows = 20 (데이터나오는 수 변경)
```

```python
# 국가별 매출(.groupby 활용)
rev_by_countries = retail.groupby('Country')['CheckoutPrice'].sum().sort_values()
rev_by_countries
```

```python
# 막대그래프 그리기
plot = rev_by_countries.plot(kind='bar', figsize=(20,10))
plot.set_xlabel('Country', fontsize=12)
plot.set_ylabel('Revenue', fontsize=12)
plot.set_title('Revenue By Country', fontsize=15)
plot.set_xticklabels(labels=rev_by_countries.index, rotation=45)
```

```python
rev_by_countries / total_revenue
```

```python
# 월별 매출 구하기
retail['InvoiceDate'].sort_values(ascending=False)
```

```python
def extract_month(date): # 2011-12-09 12:50:00
    month = str(date.month) # 월 추출 12
    if date.month < 10:
        month = '0' + month # 02로 만들어줌
    return str(date.year) + month # 201112, 201109로 연도+월 리턴가능
```

```python
rev_by_month = retail.set_index('InvoiceDate').groupby(extract_month)['CheckoutPrice'].sum()
rev_by_month
# set_index : 인덱스로 세팅한다 / 콜백함수 사용
```

```python
# 막대그래프 그리기
def plot_bar(df, xlabel, ylabel, title, rotation=45, titlesize=15, fontsize=12, figsize=(20,10)):
    plot = df.plot(kind='bar', figsize=figsize)
    plot.set_xlabel(xlabel, fontsize=fontsize)
    plot.set_ylabel(ylabel, fontsize=fontsize)
    plot.set_title(title, fontsize=titlesize)
    plot.set_xticklabels(labels=df.index, rotation=rotation)
```

```python
plot_bar(rev_by_month,'Month','Revenue','Revenue By Month')
```

```python
# 요일별 매출 구하기
```

```python
def extract_day(date): # 2011-12-09 12:50:00
    day = str(date.dayofweek) # 월 추출 12

    return day # 201112, 201109로 연도+월 리턴가능
```

```python
rev_by_dow = retail.set_index('InvoiceDate').groupby(extract_day)['CheckoutPrice'].sum()
rev_by_dow
rev_by_dow.index
#rev_by_dayofweek = retail.set_index('InvoiceDate').groupby(lambda date:date.dayofweek)['CheckoutPrice'].sum() 람다함수로 사용가능
```

```python
# 시간대별 매출 구하기
rev_by_hour = retail.set_index('InvoiceDate').groupby(lambda date: date.hour)['CheckoutPrice'].sum()
rev_by_hour
```
